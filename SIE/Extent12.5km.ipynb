```python
import matplotlib.pyplot as plt
import numpy as np
import xarray as xr
%matplotlib inline
```

```python
ds =  xr.open_dataarray('/Volumes/WorkDrive/melt_dates/SIC_07132012_02232023.nc')
area = xr.open_dataset('/Volumes/WorkDrive/melt_dates/NSIDC0771_CellArea_PS_S12.5km_v1.0.nc')
```

```python
latitude = ds.GridLat_SpPolarGrid12km
longitude = ds.GridLon_SpPolarGrid12km
time = ds.time
```

```python
sic_xr = ds.where(ds < 101,np.nan)
```

```python
sic = sic_xr.values
cell = area.cell_area
cell = cell.values
#cell = cell/1000
```

```python
sic = sic * 100 
```

```python
#cell = cell/1e8
```

```python
seaice_mask = sic >=15
```

```python
seaice_mask.shape
```

```python
sic.shape
```

```python
cell.shape
```

```python
extent = seaice_mask * cell * sic
```

```python
extent = extent/100000
```

```python
extent.shape
```

```python
da = xr.DataArray(
    data=extent,
    dims=["time","ygrid","xgrid"],
    coords=dict(
        time=time,
        longitude=(["ygrid","xgrid"],longitude),
        latitude=(["ygrid","xgrid"],latitude),
        ),
        attrs=dict(
        description="Sea Ice Extent",
        units="km2",
        ),
)
```
